## PYTHON OPERATOR

>In Python, operators are special symbols, combinations of symbols, or keywords that designate some type of computation. You can combine objects and operators to build expressions that perform the actual computation. So, operators are the building blocks of expressions, which you can use to manipulate your data. Therefore, understanding how operators work in Python is essential for you as a programmer.
>
>You’ll find several categories or groups of operators in Python. Here’s a quick list of those categories:
>
>Assignment operators
>
>Arithmetic operators
>
>Comparison operators
>
>Boolean or logical operators

```python
# arithmetic operator example
a=5
b=2
```

```python
# addition (sum)
print("a + b = ", a + b)

# substraction
print("a - b = ", a - b)

# multiplication
print("a * b = ", a * b)

# division
print("a / b = ", a / b)

#floor division
print("a // b = ", a // b)

#modulo
print("a % b = ", a % b)

#exponent (a to the power b)
print("a ** b = ", a ** b)
```

### ASSIGNMENT

```python
# assignment operator example
a=1
b=2
```

```python
# assign the sum of a and b to a
a+=b
print("Values of'a' is", a)
print("Values of'b' is", b)
```

```python
# substraction
a -= b
print ("a-b= ", a)
```

```python
#multiplication
a *= b
print("a * b = ", a)
```

```python
# division
a /= b
print("a / b= ", a)
```

```python
# exponent
a **= b
print("a ** b =", a)
```

### COMPARISON

```python
# comparison operator example
a=5
b=2
```

```python
# less than
print("'a' is less than 'b' =", a< b)
```

```python
# less than or equal to
print("'a' is less than or equal to 'b'=", a<=b)
```

```python
# greater than
print("'a' is greater than 'b' =", a>=b)
```

```python
# greater than or equal to
print("'a' is greater than or equal to 'b'=", a>=b)
```

```python
# equal to
print("'a' is equal than 'b' =", a==b)
```

```python
# not equal to
print("'a' is not equal than 'b' =", a!=b)
```

### LOGICAL(BOOLEAN)

```python
# logical operator example
a = True
b = False
```

```python
# and
c = a and b
print (a,b,c)
```

```python
# or
c = a or b
print (a,b,c)
```

```python
# not
c = not a
print(a,c)
```

```python
# another example
print(5 < 7 and 3 == 3)
print(5 < 7 and 3 != 3)
print(5 != 7 and 3 == 3)
```
