### Python For And While Loop

>### Python Iterators
>An iterator is an object that contains a countable number of values.
>
>An iterator is an object that can be iterated upon, meaning that you can traverse through all the values.
>
>Lists, tuples, dictionaries, and sets are all iterable objects. They are iterable containers which you can get an iterator from.

### For
Python For loop is used for sequential traversal i.e. it is used for iterating over an iterable like String, Tuple, List, Set, or Dictionary.
[]

```python
# Example of a simple for loop
x = ['a','b','c','d']
for val in x:
    print(val)
```

```python
# example looping for string
string = 'Hello World'
for char in string:
    print(char)
```

>The range() function returns a sequence of numbers, starting from 0 by default, and increments by 1 (by default), and ends at a specified number.

```python
# Example of range loop with range()
# (start,stop,step)
for i in range (3,10,3):
    print(f'Number {i}')
```

>### FOR LOOP AND ELSE
>The else keyword in a for loop specifies a block of code to be executed when the loop is finished
>
>The else block will NOT be executed if the loop is stopped by a break statement.

```python
# Example of for loop with else
for i in range(5):
    if i == 4:
        print('Hello')
    else:
        print(f'Number {i}')
else:
    print('No more numbers.')
```

>### While Loop
>With the while loop we can execute a set of statements as long as a condition is true.
>
>The while loop requires relevant variables to be ready, in this example we need to define an indexing variable, i, which we set to 1.

```python
# example of a simple while loop
count = 0
while count < 5:
    print(f'Count: {count}')
    count += 1
```

```python
# example with else while loop
count = 0
while count < 6:
    if count == 2:
        print(f'Hello')
        count += 1
    else:
        print(f'Count: {count}')
        count += 1
else:
    print('No more count.')
```

```python
# sum of numbers in a list using for loop
numbers = [1,2,3,4,5]
total = 0
for num in numbers:
    total += num
    print(f'Total = {total}')
print(f'The sum of numbers in the list is: {total}')
```

```python
count = 5
while count > 0:
    print(f'{count}')
    count -= 1
print('Close the door')
```

```python
# iterating through a dictionary using for loop
person = {'name': 'Jhon',
         'age': 30,
         'city': 'New York'}
for key, value in person.items():
    print(f'{key}:{value}')
```

```python
# prime number Chechker using for loop
num = 29 # change this value to test other numbers
is_prime = True
for i in range(2, int(num ** 0.5) + 1):
    if num %i == 0:
        is_prime = False
        break
if is_prime and num != 1:
    print(f'{num} is a prime number')
else:
    print(f'{num} is not a prime number')
```

>The `zip()` function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together etc.

```python
# exampe of zip()
names = ['Alice','Bob','Charlie']
ages = [25, 30, 35]
for name, age in zip(names, ages):
    print(f'{name} is {age} years old.')
```

>The enumerate() function takes a collection (e.g. a tuple) and returns it as an enumerate object.
>
>The `enumerate()` function adds a counter as the key of the enumerate object.

```python
# example of enumerate()
fruits = ['apple', 'banana', 'cherry']
for idx, val in enumerate(fruits):
    print(f'Fruit {idx +1}: {val}')
```

>### break 
>To break out of a loop

```python
for i in range(9):
    if i > 3:
        break
    print(i)
```

```python
i = 1
while i < 9:
    print(i)
    if i == 3:
        break
    i += 1
```

>### Continue
>To continue to the next iteration of a loop

```python
for i in range(9):
    if i == 5:
        continue
    print(i)
```

```python
i = 0
while i < 9:
    i += 1
    if i == 3:
        continue
    print(i)
```

>### Nested Loop
>A nested loop is a loop inside a loop.
>
>The "inner loop" will be executed one time for each iteration of the "outer loop"

```python
# example of nested loop
adj = ["red", "big", "tasty"]
fruits = ["apple", "banana", "cherry"]

for i in adj:
    for j in fruits:
        print(i, j)
```

```python
# other example
for i in range(5):
    for j in range(i + 1):
        print(j, end='')
    print()
```

>use nested loop to find common elements in two list

```python
#finding common element in 2
list1 = [1,2,3,4,5]
list2 = [4,5,6,7,8]
common_elements = []
for element1 in list1:
    for element2 in list2:
        if element1 == element2:
            common_elements.append(element1)
```

```python
#printing result
print('Common element:', common_elements)
```
